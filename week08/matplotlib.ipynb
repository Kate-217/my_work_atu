### Write a program that makes a list called ages that has, the same number random values as salaries, (range:21 to 65) . Make scatter plot of this data.

```python

min_age = 21
max_age = 65
number_of_entries = 10
ages = np.random.randint(min_age, max_age,number_of_entries)
np.random.seed(1) 
ages
```

#### Write program that has a list of counties, make it a long list (pick from five counties). 
#### Make some counties appear more than others. Make a pie chart of the counties.

```python
possible_counties = ['Mayo', 'Galway','Roscommon','Dublin','Donegal', 'Kerry','Kilkenny']
counties = np.random.choice(possible_counties, p = [0.1, 0.02, 0.06, 0.12, 0.08, 0.3, 0.32],size = 100)
unique, counts = np.unique(counties, return_counts=True)
plt.pie(counts, labels=unique, autopct='%1.1f%%')
plt.title('pie chart of counties')
plt.show()

```

```python
plt.bar(unique, counts)
plt.show()
```
